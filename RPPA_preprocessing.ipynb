# Preprocessing data for figure 4a

The code presenteed below allows to create the dataset used for reproducing **Figure 4a**.

```python
# read data with rppa data
data_sample = pd.read_csv('data/data_clinical_sample.txt',delimiter='\t',header=2,skiprows=range(0, 2))

data_sample.head(2)
```

```python
#selecting the columns of interest.
data_sample=data_sample[['PATIENT_ID','SAMPLE_ID','CANCER_TYPE_DETAILED','ER_STATUS_BY_IHC','PR_STATUS_BY_IHC','IHC_HER2','ONCOTREE_CODE']]
```

```python
#select only rows containing ILC and IDC 
data_IDC=data_sample[data_sample.ONCOTREE_CODE=='IDC']
data_ILC=data_sample[data_sample.ONCOTREE_CODE=='ILC']
```

## Part 1: Selection of luminal A IDC and luminal A ILC samples

```python
# read data with rppa data/ mmc2
data_A = pd.read_excel('data/lumA.xlsx',header=1,skiprows=range(0, 1))

data_A.head(2)
```

```python
#selectiong only luminal A samples from IDC and ILc
data_A=data_A[data_A['PAM50']== 'LumA']
data_A.head(2)
```

```python
#data with boh IDC and ILC samples that are luminal A 
data_both=data_A[(data_A['Final Pathology']=='IDC') | (data_A['Final Pathology']=='ILC')]
```

```python
#data with seperate histological subtype
data_A_IDC=data_A[data_A['Final Pathology']=='IDC']
data_A_ILC=data_A[data_A['Final Pathology']=='ILC']
```

```python
#patient id with IDC
selected_patients_IDC=data_A_IDC['Case.ID']
selected_patients_IDC
```

```python
#patient id with ILC
selected_patients_ILC=data_A_ILC['Case.ID']
selected_patients_ILC
```

```python
#patient id with IDc or ILC
selected_patients_both=data_both['Case.ID']
selected_patients_both
```

```python
#keep only patient that are luminal A 
data_IDC=data_IDC[data_IDC.PATIENT_ID.isin(selected_patients_IDC)]
data_ILC=data_ILC[data_ILC.PATIENT_ID.isin(selected_patients_ILC)]
```

```python
data_both=data_sample[data_sample.PATIENT_ID.isin(selected_patients_both)]
```

```python
#keep the column of interest
data_both=data_both[['SAMPLE_ID','ONCOTREE_CODE']]
data_both
```

```python
data_both.rename(columns={"SAMPLE_ID": "sample_id", "ONCOTREE_CODE": "group"},inplace=True)
```

```python
#sample id of patient with ILC or IDC 
selected_ILC=data_ILC['SAMPLE_ID']
selected_IDC=data_IDC['SAMPLE_ID']
```

## Part 2: Selecting only ILC and IDC samples fron RPPA data

```python
# read data with rppa data
data_rppa = pd.read_csv('data2/data_rppa.txt',delimiter='\t')
data_rppa.head(2)
```

```python
#only keep column where the sample is IDC  and add protein columns
rppa_IDC = data_rppa[data_rppa.columns.intersection(selected_IDC)]
rppa_IDC['protein']=data_rppa['Composite.Element.REF']
```

```python
#only keep column where the sample is ILC and add protein columns
rppa_ILC = data_rppa[data_rppa.columns.intersection(selected_ILC)]
rppa_ILC['protein']=data_rppa['Composite.Element.REF']
```

```python
#keep only the protein name

new=rppa_ILC["protein"].str.split("|", n = 1, expand = True) 
rppa_ILC["protein"]=new[1]

new=rppa_IDC["protein"].str.split("|", n = 1, expand = True) 
rppa_IDC["protein"]=new[1]
```

```python
#protein of interest
protein_list=['STAT3_pY705','PI3K-p110-alpha','PI3K-p85','PTEN','INPP4B','Akt','Akt_pS473','Akt_pT308','EGFR','EGFR_pY1068','EGFR_pY1173','mTOR_pS2448','p70S6K_pT389','PI3K']

```

## Part 3: columns reordering

```python
#keep the protein of interest
rppa_ILC=rppa_ILC[rppa_ILC.protein.isin(protein_list)]
rppa_IDC=rppa_IDC[rppa_IDC.protein.isin(protein_list)]
```

```python
#create a list of protein
protein=rppa_IDC['protein']
```

```python
rppa_IDC.drop(columns=['protein'],inplace=True)
```

```python
rppa_IDC.loc[0]=list(rppa_IDC.mean(axis=0))
#rppa_IDC.mean(axis=0)
```

```python
rppa_IDC=rppa_IDC.T
```

```python
rppa_IDC=rppa_IDC.sort_values(by=[0])
```

```python
rppa_IDC.drop(columns=[0],inplace=True)
```

```python
rppa_IDC=rppa_IDC.T
```

```python
rppa_IDC['protein']=protein
```

```python
rppa_ILC.drop(columns=['protein'],inplace=True)
rppa_ILC.loc[0]=list(rppa_ILC.mean(axis=0))
rppa_ILC=rppa_ILC.T
```

```python
rppa_ILC=rppa_ILC.sort_values(by=[0])
rppa_ILC.drop(columns=[0],inplace=True)
rppa_ILC=rppa_ILC.T
rppa_ILC['protein']=protein
```

## Part 4: Final data with both ILC and IDC samples in the right order

```python
#final rppa data
df= pd.concat([rppa_IDC,rppa_ILC], axis=1)
```

```python
lista=list(df.columns)
```

```python
data_both=data_both[data_both['sample_id'].isin(lista)]
```
