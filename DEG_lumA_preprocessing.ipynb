```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
sns.set()
```

This code shows how we extracted lumA IDC and lumA IDC samples for part 4.2.4 of the project: **Comparison between lumA ILC and lumA IDC**.

## 1-  Extraction of the luminal A patients  (IDC & ILC)

```python
# read data (tableS1 from the article)
data_A = pd.read_excel('data/lumA.xlsx',header=1,skiprows=range(0, 1))

data_A.head(3)
```

```python
#Number of IDC and ILC samples
data_A.groupby('Final Pathology').count()
```

```python
#keep only the LumA type
data_A=data_A[data_A['PAM50']== 'LumA']

#keep only IDC and ILC breast cancer
cancer_type=['IDC','ILC']
data_A=data_A[data_A['Final Pathology'].isin(cancer_type)]

data_A.head(3)
```

```python
#selection of the case.id (patient id) that are luminal A IDC or luminal A ILC

selected_patients=data_A['Case.ID']
selected_patients
```

## 2- Extraction of the luminal A samples ID (IDC & ILC)

**AIM**: extract the samples ID with luminal A ILC or luminal A IDC.

```python
# read data with clinical sample data (link between patient id and sample id)
data_sample = pd.read_csv('data/data_clinical_sample.txt',delimiter='\t',header=2,skiprows=range(0, 2))

data_sample.head(2)
```

```python
#select the column of interest
data_sample=data_sample[['PATIENT_ID','SAMPLE_ID','CANCER_TYPE_DETAILED','ER_STATUS_BY_IHC','PR_STATUS_BY_IHC','IHC_HER2','ONCOTREE_CODE']]
```

```python
#keep only the data with IDC and ILC 
data=data_sample[data_sample.ONCOTREE_CODE.isin(cancer_type)]
```

```python
#keep only the patient with a luminal A cancer type
data=data[data.PATIENT_ID.isin(selected_patients)]
```

```python
#list with the sample id of interest
selected_samples=data['SAMPLE_ID']
```

## 3- Genes expression of each selected lumA sample (IDC or ILC)

**AIM**: Filter the RNA-seq data in order to keep only samples that are classified as luminal A ILC or luminal A IDC.

```python
# read data with rppa data
data_seq = pd.read_csv('data2/data_RNA_Seq_v2_expression_median.txt',delimiter='\t')
data_seq.head(2)
```

```python
#new dataframe with only the lumA+ samples
data_seq = data_seq[data_seq.columns.intersection(selected_samples)]

data_seq.head(3)
```
